```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import mpl_toolkits.mplot3d
```

```python
df=pd.read_csv('/content/drive/MyDrive/House Price India.csv')
```

```python
df.head()
```

```python
df.shape
```

Univariate Analysis  

a) Histogram:


```python
plt.hist(['Renovation Year'], bins=10)
plt.title('Histogram of Variable')
plt.xlabel('bulding years')
plt.ylabel('Frequency')
plt.show()
```

B.box plot

```python
plt.boxplot(df['number of bedrooms'])
plt.title('Box Plot of Variable')
plt.ylabel('Variable Values')
plt.show()
```

```python
df_01=df.loc[df['Built Year']==2000]
df_55=df.loc[df['Built Year']==1955]
df_69=df.loc[df['Built Year']==1969]
```

```python
plt.plot(df_01['id'],np.zeros_like(df_01['id']),'o')
plt.plot(df_55['id'],np.zeros_like(df_55['id']),'o')
plt.plot(df_69['id'],np.zeros_like(df_69['id']),'o')
plt.xlabel('id')
plt.show()
```

Bivariate Analysis

a) Scatter Plot:

```python
plt.scatter(df['Date'], df['living area'])
plt.title('Scatter Plot of Variables')  
plt.xlabel('Date')
plt.ylabel('living area')
plt.show()
     
```

b) Bar Plot:

```python
sales_by_region = df.groupby(['Number of schools nearby', 'Distance from the airport'])['Price'].sum().reset_index()

plt.bar(sales_by_region['Number of schools nearby'].astype(str)  + ' - ' + sales_by_region['Distance from the airport'].astype(str) , sales_by_region['Price'])
plt.title('Total price depends on Number of schools nearby Distance from the airport')
plt.xlabel('Number of schools nearby Distance from the airport')
plt.ylabel('Price')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.FacetGrid(df,hue="number of bedrooms",height=5).map(plt.scatter,"Built Year","number of floors").add_legend();
plt.show()
     
```

Multivariate Analysis

a) Heat Map:

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Heat Map of Correlation Matrix')
plt.show()
```

b) 3D Scatter Plot:

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['living area'], df['Date'], df['number of floors'])
ax.set_xlabel('Variable 1 Values')
ax.set_ylabel('Variable 2 Values')
plt.title('3D Scatter Plot of Variables')
plt.show()
```
